```python
#optimizer ---> taking root by importing, x+cos(x) if x=0 => 0+root of(x)
#cant get answer in maths so we use scipi root --> it will predict the value and give , will show x value so res.x => gives the answer

from scipy.optimize import root
from math import cos
def equation(x):
  return x+cos(x)
res=root(equation,0)
print(res.x)


```

```python
#CSR COMPRESSED SPARSE ROW
import numpy as np
from scipy.sparse import csr_matrix
a=np.array([5,0,0,8,0,9,0,0,0,1,0])# in default it will consider as 2 dim(will consider whole as zero)
print(csr_matrix(a))# non zero values with index will be shown
```

```python
# csr matrix data property --> non zero values will be given in arrays
import numpy as np
from scipy.sparse import csr_matrix
a=np.array([5,0,0,8,0,9,0,0,1,0])
print(csr_matrix(a).data)
```

```python
#count_non zero()
import numpy as np
from scipy.sparse import csr_matrix
a=np.array([5,6,7,8,9,0,0,1,0])
print(csr_matrix(a).count_nonzero())
```

```python
#standard deviation

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
a=np.array([2,3,4,5,55,6,7,88,56,55,])
sd=np.std(a)
print(np.mean(a))
print(sd)
sb.distplot(a,hist=False)
plt.show()
```

```python
#variance

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
a=np.array([50,60,80,90,70,60,12,8])
v=np.var(a)
print(np.std(a))
print(v)
sb.distplot(a,hist=False)
plt.show()
```
